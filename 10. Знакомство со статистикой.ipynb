Сразу в голову приходят минимум, максимум и среднее. Бывают ещё мода и медиана. Давайте выясним, что это такое.

*Минимум* -- это наименьшее число в массиве.

*Максимум* -- это наибольшее число в массиве.

*Среднее арифметическое* -- это сумма всех чисел в массиве, делённая на их количество.

*Мода* -- это самое часто встречающееся число в массиве.

*Медиана* вычисляется так. Упорядочим все числа в массиве по возрастанию. Если чисел в массиве нечётное количество, возьмём то, которое стоит посередине. Если чисел чётное количество, возьмём среднее арифметическое двух чисел посередине.

Такие характеристики набора данных называются **статистики**.
